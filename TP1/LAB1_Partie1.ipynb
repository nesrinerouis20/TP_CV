## 1) Image Augmentation with OpenCV

### 1.1 Installation and Setup


```python
import cv2
```

## 1.2 Loading and Displaying Image 

```python
image = cv2.imread('bird.jpg')
cv2.imshow("bird",image)
cv2.waitKey(0)
cv2.destroyAllWindows() 
```

## 1.3 Image Transformations

```python
import numpy as np
from matplotlib import pyplot as plt
def image_augmentation(image,path):
#Fliping   
    im1=cv2.flip(image,0)
    cv2.imwrite(path + '/im1.jpg',im1)
#cropping
    im2=image[80:640,90:399]
    cv2.imwrite(path + '/im2.jpg',im2)
# Scaling
    scale_percent = 10 # percent of original size
    width = int(image.shape[1] * scale_percent / 100)
    height = int(image.shape[0] * scale_percent / 100)
    dim = (width, height)
    im3=cv2.resize(image,dim)
    cv2.imwrite(path + '/im3.jpg',im3)
# Rotating:
    im4=cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
    cv2.imwrite(path + '/im4.jpg',im4)
# Brightness
    im5 = np.ones(image.shape, dtype = "uint8") * 120 
    image_bri = cv2.add(image, im5)
    cv2.imwrite(path + '/im5.jpg',image_bri)
#Median Blur
    im6=cv2.medianBlur(image,5)
    cv2.imwrite(path + '/im6.jpg',im6)
#Sharpening:
    im7= np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
    sharpened = cv2.filter2D(image, -1, 5)
    cv2.imwrite(path + '/im7.jpg',sharpened)
    
    plt.subplot(2,4,1), plt.imshow(image), plt.title("original")
    plt.subplot(2,4,2), plt.imshow(im1), plt.title("Flip")
    plt.subplot(2,4,3), plt.imshow(im2), plt.title("cropping")
    plt.subplot(2,4,4), plt.imshow(im3), plt.title("Scaling")
    plt.subplot(2,4,5), plt.imshow(im4), plt.title("Rotating")
    plt.subplot(2,4,6), plt.imshow(im5), plt.title("Brightness")
    plt.subplot(2,4,7), plt.imshow(im6), plt.title("Median Blur")
    plt.subplot(2,4,8), plt.imshow(im7), plt.title("Sharpening")
```
